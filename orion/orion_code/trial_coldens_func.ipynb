```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
x = np.random.normal(3, 1, 1000)
y = np.random.normal(4, 1, 1000)

# Create a 2D histogram
fig, ax = plt.subplots()
hist, xedges, yedges, im = ax.hist2d(x, y, bins=(50, 50), cmap=plt.cm.viridis)

# Select a range of x values
x_range = (-1, 5)
x_indices = np.where((xedges >= x_range[0]) & (xedges <= x_range[1]))[0]

# Compute the median of the y values over the selected x range
y_values = []
for i in x_indices:
    y_bin = hist[i, :]
    y_values.extend(y_bin[y_bin > 0])
median_y = np.median(y_values)

print("Median of y values over x range {} = {:.2f}".format(x_range, median_y))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Create a 2D histogram
fig, ax = plt.subplots()
hist, xedges, yedges, im = ax.hist2d(x, y, bins=50, cmap=plt.cm.viridis)

# Select a range of x values
x_range = (-1, 1)
x_indices = np.where((xedges >= x_range[0]) & (xedges <= x_range[1]))[0]

# Compute the median of the y values over the selected x range
y_values = []
for i in x_indices:
    y_bin = hist[i, :]
    y_values.extend(y_bin[y_bin > 0])
median_y = np.median(y_values)

# Plot the median as a function of x
x_centers = (xedges[:-1] + xedges[1:]) / 2
median_y_values = []
for i in range(len(x_centers)):
    x_center = x_centers[i]
    x_bin = hist[i, :]
    y_bin_indices = np.where((x >= xedges[i]) & (x < xedges[i+1]))[0]
    if len(y_bin_indices) > 0:
        y_bin = y[y_bin_indices]
        median_y_bin = np.median(y_bin)
        median_y_values.append(median_y_bin)
    else:
        median_y_values.append(np.nan)
ax.plot(x_centers, median_y_values, 'w--', lw=2)
ax.text(x_range[1]-0.5, median_y+0.1, 'Median = {:.2f}'.format(median_y), color='white', ha='right')

plt.show()
```
