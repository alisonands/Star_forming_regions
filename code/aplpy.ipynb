```python
import aplpy
import numpy as np
```

```python
gc = aplpy.FITSFigure('tutorial/fits/2MASS_k.fits') #create canvas for plotting figure
gc.show_grayscale()                                 #display image using grayscale stretch
gc.show_colorscale()                                #colourscale image
gc.show_colorscale(cmap = 'gist_heat')              #'heat' map
gc.show_rgb('tutorial/graphics/2MASS_arcsinh_color.png')
gc.tick_labels.set_font(size='large')
```

```python
gc = aplpy.FITSFigure('tutorial/fits/2MASS_k.fits') #create canvas for plotting figure
gc.show_rgb('tutorial/graphics/2MASS_arcsinh_color.png')

gc.show_contour('tutorial/fits/mips_24micron.fits', colors='white')
gc.add_grid()
gc.show_markers(ra, dec, edgecolor = 'lightblue', facecolor='none', marker='o', s=10, alpha=0.5)
gc.list_layers()
gc.save('myfirstplot.png')
```
